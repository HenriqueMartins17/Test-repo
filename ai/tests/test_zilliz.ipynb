# Test Zilliz Cloud


Script that to test zilliz vector embedding.

```python
from ai_server.trainers.base_trainer import BaseTrainer
import os
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Zilliz
from langchain.document_loaders import WebBaseLoader
from langchain.text_splitter import CharacterTextSplitter
import os

def test_empty_trainer():
    loader = WebBaseLoader([
        "https://milvus.io/docs/overview.md",
    ])
    docs = loader.load()
    # Split the documents into smaller chunks
    text_splitter = CharacterTextSplitter(chunk_size=1024, chunk_overlap=0)
    docs = text_splitter.split_documents(docs)
    # Use the WebBaseLoader to load specified web pages into documents
    # Set up an embedding model to covert documents into vector embeddings
    embeddings = OpenAIEmbeddings(model="ada")

    # Set up a vector store used to save the vector embeddings. Here we use Milvus as the vector store
    vector_store = Zilliz.from_documents(
        docs,
        embedding=embeddings,
        collection_name="LangChainCollection2",
        connection_args={"uri": ZILLIZ_ENDPOINT, "token": ZILLIZ_TOKEN, "secure": True}
    ) 
def test_empty_query():
    embeddings = OpenAIEmbeddings(model="ada")
    vector_store = Zilliz(embeddings,
        collection_name="LangChainCollection2",
        connection_args={"uri": ZILLIZ_ENDPOINT, "token": ZILLIZ_TOKEN, "secure": True}
    )
    query = "What is milvus?"
    docs = vector_store.similarity_search(query)
    print(docs)

if __name__ == '__main__':
    from dotenv import load_dotenv
    load_dotenv('../')
    ZILLIZ_ENDPOINT = os.getenv("ZILLIZ_ENDPOINT")
    ZILLIZ_TOKEN = os.getenv('ZILLIZ_TOKEN')
    OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')

    test_empty_trainer()
    test_empty_query()
```
